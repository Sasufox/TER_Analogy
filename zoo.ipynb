```python
def exist_comb(a,b,c):
    return (a==b) or (a==c)

print(exist_comb(0,1,1))
print(exist_comb(1,1,1))
print(exist_comb(0,0,0))
print(exist_comb(1,0,1))
```

```python
def create_d(a,b,c):
    if a==b : return c
    else : return b

print(create_d(1,1,1))
print(create_d(0,0,0))
print(create_d(1,0,1))
print(create_d(0,1,0))
```

```python
def create_comb(liste):
    liste_final=[]
    l1 = liste[0]
    l2 = liste[1]
    l3 = liste[2]
    for i in range(1,len(l1)):
        #print("len l + ", len(l))
        a=l1[i]
        b=l2[i]
        c=l3[i]
        if exist_comb(a,b,c):
            liste_final.append(create_d(a,b,c))
    if len(liste_final)==len(l1)-1 : return liste_final
    return None
```

```python
from itertools import chain, combinations

# powerset
def powerset(iterable):
    "powerset([1,2,3]) â†’ () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return tuple_list_into_list_list(chain.from_iterable(combinations(s, r) for r in range(2, 4))) # len(s)+1
#list(powerset(resultat[0][0]))
#print(len(powerset(resultat[0][0])), len((resultat[0])), len((resultat)))
```

```python
import numpy as np
from scipy import spatial
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

print("Loading Glove Model")
f = open("glove.840B.300d.txt",'r', encoding='utf8')
embeddings_dict = {}
for line in f:
    splitLine = line.split(' ')
    word = splitLine[0]
    embedding = np.asarray(splitLine[1:], dtype='float32')
    embeddings_dict[word] = embedding
print("Done.",len(embeddings_dict)," words loaded!")
```

```python
def find_closest_embeddings(embedding):
    return sorted(embeddings_dict.keys(), key=lambda word: spatial.distance.euclidean(embeddings_dict[word], embedding))
```

```python
print(find_closest_embeddings(
    embeddings_dict["aardvark"] + embeddings_dict["antelope"] + embeddings_dict["bear"]
)[:10])
```

```python
def get_find_closest_embeddings(a,b,c):
    fce=find_closest_embeddings(
    embeddings_dict[a] + embeddings_dict[b] + embeddings_dict[c]
    )[:10]
    fce=[x for x in fce if x!=a and x!=b and x!=c]
    return fce

get_find_closest_embeddings("buffalo", "calf", "carp")
```
